```python
# Step 1: Convert timestamp columns into datetime variables
timestamp_columns = ['REPORT_DAT', 'START_DATE', 'END_DATE']
for col in timestamp_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python

# Step 2: Remove unnecessary columns
columns_to_drop = ['XBLOCK', 'YBLOCK', 'OCTO_RECORD_ID', 'OBJECTID']
data = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
# Step 3: Calculate the percentage of missing values for each column
missing_percentage = data.isnull().mean() * 100
print(missing_percentage)
```

```python
# Step 4: Handle missing values based on their percentage
# Separate columns by missing value thresholds
less_than_5 = missing_percentage[missing_percentage < 5].index
between_5_and_25 = missing_percentage[(missing_percentage >= 5) & (missing_percentage <= 25)].index
more_than_25 = missing_percentage[missing_percentage > 25].index
```

```python
# (a) Remove rows with missing values in columns where <5% of rows are missing
data = data.dropna(subset=less_than_5, how='any')
```

```python
# (b) Impute missing values for columns with 5%-25% missing
for col in between_5_and_25:
    if data[col].dtype in ['float64', 'int64']:  # Numerical columns
        data[col].fillna(data[col].median(), inplace=True)
    else:  # Categorical columns
        data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
# (c) Impute missing values for columns with more than 25% missing
for col in more_than_25:
    if data[col].dtype in ['float64', 'int64']:  # Numerical columns
        data[col].fillna(data[col].median(), inplace=True)
    else:  # Categorical columns
        data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
# Step 5: Remove rows where 'END_DATE' is missing
data = data.dropna(subset=['END_DATE'], how='any')
```

```python
df['REPORT_DAT'] = pd.to_datetime(df['REPORT_DAT'])

# Extract month and week from the REPORT_DAT column
df['Month'] = df['REPORT_DAT'].dt.month
df['Week'] = df['REPORT_DAT'].dt.isocalendar().week

# Summarize data for monthly and weekly counts
monthly_counts = df.groupby('Month').size().reset_index(name='Counts')
weekly_counts = df.groupby('Week').size().reset_index(name='Counts')

# Display the summaries
print("Monthly Counts:")
print(monthly_counts)

print("\nWeekly Counts:")
print(weekly_counts)
```

```python
# Ensure the REPORT_DAT column is in datetime format
df['REPORT_DAT'] = pd.to_datetime(df['REPORT_DAT'])

# Define a function to map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

# Apply the function to create a Seasons column
df['Season'] = df['REPORT_DAT'].dt.month.apply(get_season)

# Display the updated dataset (first few rows)
print(df[['REPORT_DAT', 'Season']].head())
```
